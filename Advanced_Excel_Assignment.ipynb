**Question 1 :** Explain the difference between Absolute, Relative, and Mixed CellReferencing  in Excel with examples.

**Answer :** **1. Absolute Cell Referencing ($A$1)**

**Meaning:**
An absolute reference does not change when you copy the formula to another cell. Both the column and row are fixed.

**Example:**
Suppose cell A1 has value 10 and cell B1 has value 5.

Formula in C1:
=A1*$B$1

If you copy this formula from C1 to C2, it becomes:
=A2*$B$1

Here, $B$1 stays the same even after copying.

**2. Relative Cell Referencing (A1)**

**Meaning:**
A relative reference changes automatically when the formula is copied to another cell.

**Example:**
Formula in C1:
=A1+B1

If you copy this formula to C2, it becomes:
=A2+B2

Both row and column adjust relative to the new position.

**3. Mixed Cell Referencing ($A1 or A$1)**

**Meaning:**
In mixed referencing, either the row or the column is fixed, but not both.

**Types and Examples:**=

$A1 (Column fixed, row changes)
Formula: =$A1*B1
When copied down, column A remains fixed but row number changes.

A$1 (Row fixed, column changes)
Formula: =A$1*B1
When copied across, row 1 remains fixed but column changes.

**Question 3 :** What are Text Functions in Excel? Mention any five with examples.

**Answer :** Text functions in Excel are used to manipulate, analyze, and format text (string) data. They help in cleaning data, combining text, extracting specific characters, and changing text formats.

**Five Text Functions in Excel with Examples**

1. LEFT()

Purpose: Extracts characters from the left side of a text.

Example:
Formula: =LEFT("Excel",2)
Result: Ex

2. RIGHT()

Purpose: Extracts characters from the right side of a text.

Example:
Formula: =RIGHT("Excel",3)
Result: cel

3. MID()

Purpose: Extracts characters from the middle of a text.

Example:
Formula: =MID("Microsoft",2,4)
Result: icro

4. LEN()

Purpose: Returns the total number of characters in a text.

Example:
Formula: =LEN("Excel")
Result: 5

5. CONCAT() (or CONCATENATE())

Purpose: Joins two or more text strings into one.

Example:
Formula: =CONCAT("Data"," ","Analysis")
Result: Data Analysis

**Question 5 :** Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?


**Answer :** **Purpose of VLOOKUP**

VLOOKUP (Vertical Lookup) is used to search for a value in the first column of a table and return a related value from another column in the same row.

**Example:**

=VLOOKUP(101, A2:D10, 3, FALSE)
Searches for 101 in column A and returns the value from the 3rd column.

**Purpose of HLOOKUP**

HLOOKUP (Horizontal Lookup) searches for a value in the first row of a table and returns a related value from another row in the same column.

**Example:**

=HLOOKUP("Sales", A1:D5, 2, FALSE)
Searches for “Sales” in the first row and returns data from the 2nd row.

How VLOOKUP and HLOOKUP differ from XLOOKUP

XLOOKUP is a modern and more flexible lookup function that replaces both VLOOKUP and HLOOKUP.

**Question 8 :** Create a table of 10 products with stock levels. Use Macros to highlight:

● Low stock (less than 10 units) in red

● Overstock (more than 50 units) in green.


**Answer : Step 1: Create the Product Stock Table**

Create this table in Excel (say from A1 to C11):

| Product ID | Product Name | Stock Level |
| ---------- | ------------ | ----------- |
| P001       | Laptop       | 8           |
| P002       | Mouse        | 55          |
| P003       | Keyboard     | 12          |
| P004       | Monitor      | 6           |
| P005       | Printer      | 48          |
| P006       | Scanner      | 60          |
| P007       | Webcam       | 9           |
| P008       | USB Drive    | 25          |
| P009       | Headphones   | 52          |
| P010       | Router       | 7           |

**Step 2: Purpose of the Macro**

The macro will:

- Highlight Low Stock (less than 10) in Red

- Highlight Overstock (more than 50) in Green

- Leave normal stock unchanged


```
import pandas as pd

# Step 1: Create the Product Stock Table as a pandas DataFrame
data = {
    'Product ID': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006', 'P007', 'P008', 'P009', 'P010'],
    'Product Name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer', 'Scanner', 'Webcam', 'USB Drive', 'Headphones', 'Router'],
    'Stock Level': [8, 55, 12, 6, 48, 60, 9, 25, 52, 7]
}
df = pd.DataFrame(data)

# Step 2: Implement the highlighting logic in Python
def highlight_stock_level(row):
    stock = row['Stock Level']
    if stock < 10:
        return 'Low Stock (Red)'
    elif stock > 50:
        return 'Overstock (Green)'
    else:
        return 'Normal Stock (No Highlight)'

df['Highlight Status'] = df.apply(highlight_stock_level, axis=1)

# Display the DataFrame with the new highlight status
print("Product Stock Table with Highlight Status:")
print(df.to_string())
```
